```python
#Imports
from pprint import pprint
import json
```

```python
with open('outputs/gizz_data.json') as json_data:
    data = json.load(json_data)
```

```python
def getAvg(feature):
    total = 0
    for item in feature:
        total += item
    return total/len(feature)
```

```python
stats = []
for album in data:
    valence = []
    energy = []
    danceability = []
    acousticness = []
    speechiness = []
    instrumentalness = []
    liveness = []
    duration = []
    tempo = []
    for track in album['tracks']:
        valence.append(track['features']['valence'])
        energy.append(track['features']['energy'])
        danceability.append(track['features']['danceability'])
        acousticness.append(track['features']['acousticness'])
        speechiness.append(track['features']['speechiness'])
        instrumentalness.append(track['features']['instrumentalness'])
        liveness.append(track['features']['liveness'])
        duration.append(track['features']['duration_ms'])
        tempo.append(track['features']['tempo'])
    stats.append({'album': album['title'], 'stats': {
                                                    'valence': getAvg(valence),
                                                    'energy': getAvg(energy),
                                                    'danceability': getAvg(danceability),
                                                    'acousticness': getAvg(acousticness),
                                                    'speechiness': getAvg(speechiness),
                                                    'instrumentalness': getAvg(instrumentalness),
                                                    'liveness': getAvg(liveness),
                                                    'duration': sum(duration) / 60000,
                                                    'tempo': getAvg(tempo)
    }})
```
